<font color="red" size="7">Dimensionality Reduction</font>

<font color="red" size="5">4.t-Distributed Stochastic Neighbor Embedding</font>

**t-Distributed Stochastic Neighbor Embedding (t-SNE)** is a non-linear dimensionality reduction technique specifically designed for visualizing high-dimensional datasets in a lower-dimensional space (usually 2D or 3D). Unlike PCA or LDA, t-SNE focuses on preserving local structure, making it ideal for visualizing complex data structures and clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

<font color = "yellow">Apply t-SNE:</font> We use TSNE from sklearn.manifold to apply t-SNE. The n_components=2 parameter reduces the data to 2 dimensions for visualization. The perplexity=30 parameter controls the balance between local and global structure.

```python
# Apply t-SNE to reduce to 2 dimensions
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
# Create a DataFrame for visualization
import pandas as pd
tsne_df = pd.DataFrame(data=X_tsne, columns=['TSNE1', 'TSNE2'])
tsne_df['target'] = y
```

```python
# Plot the results
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='TSNE1', y='TSNE2', hue='target', palette='Set1', data=tsne_df,
    s=100, edgecolor='k'
)
plt.title('t-SNE of Wine Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(title='Wine Class')
plt.grid()
plt.show()
```

**Key Outputs**

<ol>
<font color = "yellow"><li>Scatter Plot:</font>
    The plot shows the transformed data in 2D space. The points are color-coded according to their class labels (wine cultivars). t-SNE will aim to preserve the local relationships between the wine samples while reducing the dimensionality.</li>

<font color = "yellow"><li>Clusters:</font>
    Ideally, the plot will show clusters of points that correspond to different wine classes. t-SNE focuses on local structures, so similar wine samples will be close together in the 2D plot.</li></ol>


**Interpretation:**

<font color = "yellow">Separation:</font> If the wine classes are well-separated in the plot, this indicates that t-SNE has successfully captured the underlying structure of the data. If there is overlap, it suggests that the classes are not easily separable in the reduced 2D space.

**Advantages of t-SNE**
<ol>
<font color = "yellow"><li>Preserves Local Structure:</font>
    t-SNE is particularly effective at preserving local relationships in the data, which can reveal clusters and groups in the dataset</li>
    
<font color = "yellow"><li>Effective for Complex Data:</font> 
    t-SNE works well for complex datasets where the relationships between features are non-linear.</li>
    
<font color = "yellow"><li>Visualization:</font> 
    It is especially useful for visualizing high-dimensional data in 2D or 3D, which is valuable for exploratory data analysis.</li></ol>

**Limitations of t-SNE**

<ol>
<font color = "yellow"><li>Computationally Expensive:</font> 
    t-SNE can be slow for large datasets, especially when the number of samples is high.</li>

<font color = "yellow"><li>Non-Deterministic:</font> 
    t-SNE involves a random initialization process, so results can vary slightly between runs. However, setting a random seed can mitigate this.</li>
    

<font color = "yellow"><li>Perplexity Tuning:</font> 
    The choice of the perplexity parameter can significantly affect the results. A low perplexity will focus on smaller neighborhoods, while a high perplexity will consider larger neighborhoods, affecting the separation of clusters.</li>
    

<font color = "yellow"><li>Loss of Global Structure:</font> 
    t-SNE is good at capturing local structure but does not preserve the global relationships well (e.g., distances between distant clusters).</li></ol>

**Applications of t-SNE**

<ol>
<font color = "yellow"><li>Exploratory Data Analysis:</font>
    t-SNE is a powerful tool for visualizing high-dimensional data, revealing hidden patterns or clusters.</li>
    
<font color = "yellow"><li>Data Compression:</font>
    t-SNE can be used to reduce data to 2D or 3D while retaining the structure, making it easier to analyze or visualize.</li>
    
<font color = "yellow"><li>Clustering:</font> 
    t-SNE is useful for clustering tasks, where you want to explore how data points naturally group together in lower dimensions.</li></ol>
