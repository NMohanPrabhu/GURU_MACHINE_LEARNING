<font color="red" size="5">1_Principle Component Analysis</font>

**Principal Component Analysis (PCA)** is a linear dimensionality reduction technique that transforms a dataset into a new coordinate system. The goal is to reduce the dimensionality of the data while preserving as much variability (information) as possible.

```python
# Import necessary libraries
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for visualization
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['target'] = y

# Plot the first two principal components
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='PC1', y='PC2', hue='target', palette='Set1', data=pca_df,
    s=100, edgecolor='k'
)
plt.title('PCA of Wine Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cultivar')
plt.grid()
plt.show()

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance ratio: {explained_variance}")
print(f"Total variance explained by first 2 components: {sum(explained_variance):.2f}")
```

**Applications of PCA:**
<ol>
   <li> <font color="Red">Dimensionality Reduction:</font>
        Reduces the number of features while retaining as much information as possible. This is especially useful for high-dimensional data, such as images, genomics, and text.</li>

<li> <font color="Red">Data Visualization:</font>
        PCA is commonly used to project high-dimensional data to 2D or 3D for visualization. This helps in detecting patterns, clusters, or outliers in the data.</li>

<li> <font color="Red">Noise Reduction:</font>
        By removing components with low variance, PCA can help filter out noise, improving the signal-to-noise ratio in the data.</li>

<li> <font color="Red">Feature Engineering:</font>
        PCA can be used to create new features (principal components) that capture the most variance in the dataset, which can improve the performance of machine learning algorithms.</li></ol>
