*Feature importance from a Decision Tree is a powerful way to evaluate the relevance of features. Decision trees compute feature importance as the reduction in impurity (e.g., Gini or entropy) contributed by each feature to the model's splits. This can be extended to ensembles like Random Forests or Gradient Boosted Trees for more robust estimates.*

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the wine dataset
wine = load_wine()
X, y = wine.data, wine.target
feature_names = wine.feature_names

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize a Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=42)

# Train the model
dt.fit(X_train, y_train)
```

```python
# Get feature importances
feature_importances = dt.feature_importances_
```

```python
# Sort features by importance
sorted_indices = np.argsort(feature_importances)[::-1]
sorted_feature_names = [feature_names[i] for i in sorted_indices]
sorted_importances = feature_importances[sorted_indices]
```

```python
# Print sorted features and their importances
print("Feature Importances (Decision Tree):")
for name, importance in zip(sorted_feature_names, sorted_importances):
    print(f"{name}: {importance:.4f}")
```

```python
# Plot feature importances
plt.figure(figsize=(12, 6))
plt.bar(sorted_feature_names, sorted_importances, color="skyblue")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance from Decision Tree")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Evaluate the model
accuracy = dt.score(X_test, y_test)
print(f"Test accuracy of the Decision Tree: {accuracy:.2f}")
```
