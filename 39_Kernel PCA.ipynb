<font color="red" size="5">11_Kernel PCA</font>

**Kernel Principal Component Analysis (Kernel PCA)** is an extension of Principal Component Analysis (PCA) that allows for the analysis of non-linearly separable data by using a kernel function. While standard PCA is based on linear transformations of the data, Kernel PCA leverages the power of kernel methods to project data into a higher-dimensional space where linear separation is possible. This allows Kernel PCA to handle more complex patterns and nonlinear relationships between the features.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import KernelPCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import pandas as pd
```

```python
# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

<font color = "yellow">Kernel PCA:</font> We apply Kernel PCA using the RBF kernel (kernel='rbf') to reduce the dataset to 2 dimensions. The parameter gamma=0.1 controls the width of the RBF kernel.

```python
# Apply Kernel PCA with RBF kernel to reduce the data to 2 dimensions
kernel_pca = KernelPCA(n_components=2, kernel='rbf', gamma=0.1)
X_kpca = kernel_pca.fit_transform(X_scaled)
```

```python
# Create a DataFrame for visualization
import pandas as pd
kpca_df = pd.DataFrame(data=X_kpca, columns=['KPCA1', 'KPCA2'])
kpca_df['target'] = y
```

```python
# Plot the results
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='KPCA1', y='KPCA2', hue='target', palette='Set1', data=kpca_df,
    s=100, edgecolor='k'
)
plt.title('Kernel PCA of Wine Dataset (RBF Kernel)')
plt.xlabel('Kernel PCA Dimension 1')
plt.ylabel('Kernel PCA Dimension 2')
plt.legend(title='Wine Class')
plt.grid()
plt.show()
```

**Key Outputs**

<ol>
<font color = "yellow"><li>Kernel PCA Scatter Plot:</font>
    The plot shows the wine samples projected into a 2D space using Kernel PCA. The points are color-coded based on their wine class. You may observe that the classes (wine types) are more distinct and separated in this new space compared to a standard PCA plot (if one were to try with linear methods).</li>

<font color = "yellow"><li>Non-linear Transformation:</font> 
    Kernel PCA effectively maps the data into a higher-dimensional space using the RBF kernel, enabling better separation of the wine samples that might be non-linearly separable in the original feature space.</ol>

**Advantages of Kernel PCA**

<ol>
<font color = "yellow"><li>Non-linear Dimensionality Reduction:</font>
    Kernel PCA allows for the reduction of dimensionality in cases where the data exhibits complex non-linear relationships that cannot be captured by PCA.</li>
<font color = "yellow"><li>Flexibility:</font> 
    The ability to choose from various kernel functions (e.g., RBF, polynomial) makes Kernel PCA highly flexible and adaptable to different types of data.</li>
<font color = "yellow"><li>Improved Separation:</font>
    It helps separate data points that are non-linearly separable, which can improve classification, clustering, and other machine learning tasks.</li></ol>

**Limitations of Kernel PCA**

<ol>
<font color = "yellow"><li>Choice of Kernel:</font> 
    The performance of Kernel PCA depends on the choice of kernel and its hyperparameters (e.g., gamma in the RBF kernel). The wrong choice can lead to poor performance.</li>
<font color = "yellow"><li>Computational Complexity:</font>
    Kernel PCA can be computationally expensive, especially for large datasets, since it involves the calculation of the kernel matrix, which scales quadratically with the number of data points.</li>
<font color = "yellow"><li>Interpretability:</font>
    Unlike standard PCA, the principal components in Kernel PCA are not necessarily easily interpretable because the data has been transformed into a higher-dimensional space.</li></ol>

**Applications of Kernel PCA**

<ol>
<font color = "yellow"><li>Image Processing:</font>
    Kernel PCA can be used to extract features from images and reduce their dimensionality, making them easier to classify or cluster.</li>
<font color = "yellow"><li>Speech and Audio Analysis:</font>
    In speech processing, Kernel PCA can help extract non-linear features from audio signals.</li>
<font color = "yellow"><li>Bioinformatics:</font>
    Kernel PCA is used in genomic data analysis, where relationships between genes or samples may not be linear.</li>
<font color = "yellow"><li>Pattern Recognition:</font>
    Kernel PCA is used in pattern recognition tasks, such as face recognition or anomaly detection, where data relationships are complex and non-linear.</li></ol>
