<font color="red" size="5">2_Singular Value Decomposition (SVD)</font>

**Singular Value Decomposition (SVD)** is a matrix factorization technique that is closely related to Principal Component Analysis (PCA). While PCA focuses on finding the principal components of a dataset, SVD decomposes the data matrix into three matrices, revealing the underlying structure of the data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**SVD Decomposition:** We use np.linalg.svd() to perform SVD. This function returns:
<ol>

<li>U: Left singular vectors (representing the projection of the original data onto new orthogonal axes).</li>
<li>S: Singular values, which tell us how much variance is explained by each component.</li>
<li>VT: Right singular vectors (equivalent to the principal components in PCA).</li>
</ol>

```python
# Perform Singular Value Decomposition (SVD)
U, S, VT = np.linalg.svd(X_scaled, full_matrices=False)

# U contains the left singular vectors (representing the original data space)
# S contains the singular values (showing the importance of each component)
# VT contains the right singular vectors (representing the reduced feature space)
```

```python
# Create a DataFrame for the first two singular vectors (U)
X_svd = U[:, :2]  # Using the first 2 components for visualization
```

```python
# Plot the results
plt.figure(figsize=(10, 7))
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title('SVD of Wine Dataset')
plt.xlabel('Singular Value 1')
plt.ylabel('Singular Value 2')
plt.colorbar(label='Wine Class')
plt.grid()
plt.show()
```

```python
# Singular values
singular_values = S[:2]
print(f"Top 2 singular values: {singular_values}")
```

**Applications of SVD**
<ol>
<font color = "yellow"><li>Dimensionality Reduction:</font>
        By keeping only the top kk singular values and their corresponding vectors, we can reduce the dataset to fewer dimensions while retaining most of the variance.</li>
<font color = "yellow"><li>Latent Semantic Analysis (LSA):</font> 
        In text mining, SVD is used for latent semantic analysis, where it helps identify underlying patterns in text data.</li>
<font color = "yellow"><li>Data Compression:</font>
        By approximating the original matrix with fewer singular values, we can achieve data compression, retaining the essential structure of the data.</li></ol>
